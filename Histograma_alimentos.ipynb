```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
alimentos = pd.read_csv('C:/Users/USER/Documents/hidrologia/alimentos.csv', sep=';', index_col=0, encoding='latin1')
alimentos.head()
```

```python
bebidas = alimentos['bebidas']
alimentos_paquetes = alimentos['alimentos de paquetes']
edades = alimentos.index

# Configurar el ancho de las barras
bar_width = 0.35

# Configurar la posición de las barras en el eje x
r1 = range(len(edades))
r2 = [x + bar_width for x in r1]

# Crear el histograma
plt.bar(r1, bebidas, color='#008B8B', width=bar_width, label='Bebidas')
plt.bar(r2, alimentos_paquetes, color='#ADD8E6', width=bar_width, label='Alimentos de Paquetes')

# Configurar las etiquetas y el título
plt.xlabel('Edades')
plt.ylabel('Cantidad')
plt.title('Consumo de Bebidas y Alimentos de Paquetes por Edades')
plt.xticks([r + bar_width / 2 for r in range(len(edades))], edades, rotation=45, ha='right')
plt.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
